## 🌱 **Git stash 練習マニュアル**

### ✅ **1. stash とは？**

* **作業中の変更を一時的に避難させる「棚」**
* **LIFO（Last In, First Out）方式** → 新しい stash が一番上
* stash した内容は番号付き（`stash@{0}`, `stash@{1}`…）で管理される

---

### ✅ **2. stash の基本コマンド**

```bash
git stash         # 作業中の変更を stash（避難）
git stash list    # stash の一覧を確認
git stash pop     # stash@{0} を適用＋stash リストから削除
git stash apply   # stash@{0} を適用（stash リストには残る）
git stash drop stash@{1}  # 特定の stash を削除
```

---

### ✅ **3. stash の番号を指定して使う**

```bash
git stash apply stash@{1}
git stash pop stash@{2}
```

👉 **複数 stash があるときに古い stash を指定して使える**

---

### ✅ **4. よくあるシナリオ**

### 📌 **(1) stash を適用して main に不要な内容が混ざった**

* **変更を捨てたい場合**

```bash
git restore ファイル名
```

👉 **最後の commit 状態に戻す**

---

### 📌 **(2) stash 自体を残したい**

* `apply` を使えば stash リストに残る
* `pop` は適用と同時に stash から削除される（1回きり）

---

### ✅ **5. stash の確認**

```bash
git stash list
```

✅ stash が積み重なっているのを確認

---

### ✅ **6. stash を安全に消す**

```bash
git stash clear
```

✅ stash 全削除（**注意：全消去するので最後にやる！**）
