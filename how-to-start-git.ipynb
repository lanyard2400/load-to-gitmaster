# 1. すでにあるフォルダをgit hubで管理
〇git init → 「下流（ローカル）からGitHubへ持ち上げよう」

## ✅ **`git init` を使うシーン**

* Phil G. が **ローカルだけで作業してた自習用フォルダ** がある
* 「これ、GitHubでも管理したいな」と思ったとき…

👉 `git init` で **フォルダを「Gitで管理する箱」に変える**

---

## 🔄 **GitHub管理になると何が変わる？**

### 🌱 **1. バージョン管理ができる**

* `git add` → `git commit` で **「ここまでの状態を保存」**
* 例）**コードを壊しても過去のコミットに戻せる（タイムマシン機能）**

```bash
git checkout コミットID
```

---

### 🌱 **2. GitHubにバックアップできる**

* ローカルPCだけだと、**壊れた／紛失した／PC移行** のとき危ない
* GitHubに `push` すれば、**クラウドに安全に保管**

---

### 🌱 **3. 別PCからも使える**

* GitHubにあげておけば、**別のPCで clone して作業再開** できる

---

### 🌱 **4. 実験用ブランチで遊べる**

* `main` を壊さずに **別ブランチで実験** → OKなら `merge`
* 自習コードも「試しに別のやり方を書いてみる」が安心してできる

---

## 📌 **つまりこういう流れ**

1. これまでの自習用ディレクトリに移動

   ```bash
   cd my_study_folder
   ```
2. Git管理開始

   ```bash
   git init
   ```
🗂️ ポイント
git init を実行すると、そのフォルダの中に .git フォルダ が作られる
　（これが Git の心臓部！）


3. 最初のコミット

   ```bash
   git add .
   git commit -m "最初の状態をコミット"
   ```
4. GitHubでリポジトリ作成
5. GitHubと接続

   ```bash
   git remote add origin git@github.com:PhilG/◯◯.git  
   
   (git branch -M main  ) masterになってるときだけ

   git push -u origin main  
  
 ★-uの意味  
 をつけない場合の動き  
push はされる  
でも「このブランチは GitHub のどのブランチとペアなのか？」を  
　Git が覚えてくれない
👉 次回 push するときも、毎回こう書かないといけない👇  
git push origin work-hiro
✅ -u をつけた場合
git push -u origin work-hiro
⇒push される、さらに Git が覚える  

「ローカルの work-hiro ＝ GitHub の origin/work-hiro」
👉 次回以降は超シンプルにできる
（origin work-hiro を書かなくていい！）


   ```

---

✅ **結論**

> 自習用フォルダを `git init` → GitHubで管理すれば、
> **タイムマシン化 & クラウドバックアップ & ブランチで実験し放題**！

---


### 初めてのgit push
`git push -u origin main`** は **GitHub との「初めての接続」** のときに必ず使う大事なコマンドだよ。

---

## ✅ **分解して解説**

```bash
git push -u origin main
```

### 🌱 **1. `git push`**

* **ローカル → GitHub にデータを送る**
* 例）コミットした内容を GitHub に反映

---

### 🌱 **2. `origin`**

* **「どこに送るか」**
* `git remote add origin URL` で設定した **リモートリポジトリの名前**
* GitHub の URL（例：`https://github.com/PhilG/myrepo.git`）を
  　**短縮して `origin` って呼んでるだけ**

---

### 🌱 **3. `main`**  ※別ブランチの時はここ変わるよ！

* **「どのブランチを送るか」**
* 今いる `main` ブランチの内容を GitHub に push


---

### 🌱 **4. `-u`（--set-upstream）**

* **「このローカルブランチと、GitHubのこのブランチをペア登録する」**
* これをつけると
  　👉 **今後 `git push` だけでOK！**（`origin main` を書かなくていい）

---

## ✅ **つまり**

> **`git push -u origin main`**
> \= **「ローカルの main を GitHub の origin/main に送って、以後はこの2つを自動的にリンクする」**

---

## ✅ **次回以降どうなる？**

* 2回目からは

```bash
git push
```

だけで同じブランチに push できる 🎉


👉 **Phil G. の状況では**

* **初回 push だから `-u` が必要だった**
* もう「セットアップ」完了してるから、**次からは `git push` だけでOK**！



---


## ✅ **　mainではなく別のブランチを GitHub に送る流れ**

### ① **ローカルでブランチを作る**

```bash
git checkout -b work-hiro
```

👉 `main` から分岐して **work-hiro** ブランチを作り、そこに移動。

---

### ② **コミットする**

```bash
git add .
git commit -m "作業開始"
```

---

### ③ **初回 push（`-u` をつける）**

```bash
git push -u origin work-hiro
```

🌟 **ここが大事**

* `-u` をつけると **「ローカルの work-hiro ↔ GitHub の work-hiro」をペア登録**
* これをやれば、**2回目以降は `git push` だけでOK**

---

### ✅ **次回以降**

* もう `work-hiro` にいる状態なら

```bash
git push
```

だけでOK（`origin work-hiro` を書かなくても勝手に送ってくれる）

---

### 💡 **ポイント**

* **ブランチを新規で push するときだけ `-u` が必要**
* **すでにリンクされたブランチ** は `git push` だけでOK

---

👉 Phil G.、**次の練習で**
1️⃣ `work-hiro` ブランチ作って
2️⃣ コミット → `git push -u origin work-hiro`

ってやってみる？





もし **間違えて外のディレクトリ（例：デスクトップ全体とか）で `git init` しちゃった**場合は、
そのフォルダの中に **「.git」フォルダ** ができてるだけだから、**それを削除すればOK** だよ ✅

---

### 🔄 **間違えて `git init` したときのリセット方法**

1️⃣ **間違えて `git init` したフォルダに移動**

```bash
cd 間違えたフォルダ
```

2️⃣ **.git フォルダを削除**

* Mac / Linux / Git Bash

```bash
rm -rf .git
```

* Windows PowerShell

```powershell
rmdir .git -Force -Recurse
```

👉 `.git` フォルダを消すと、そのフォルダは **Git管理されてない普通のフォルダ** に戻る。

---

### 📌 **確認する方法**

* `ls -a`（Mac/Linux/Git Bash）や `dir /a`（Windows）で **.git があるか確認**
* 消した後、もう一度確認して `.git` がなければOK！

---

✅ **まとめ**

> `git init` の失敗＝ **.git フォルダが変な場所にできた** だけなので、
> **.git を消す＝Git管理解除** でリセットできる！
